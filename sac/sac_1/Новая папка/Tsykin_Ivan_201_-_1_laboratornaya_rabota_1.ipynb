<h2>Задание 1.2</h2>


```python
y = 2*log((x - 1) / x) + 1
show("Функция:\t", y)
```

<h3>1. Область определения:</h3>

```python
f1 = (x-1)/x
f2 = x
show(solve(f1 > 0, x) + solve(f2 != 0, x))
```

<h3>2.1. Четность функции:</h3>

```python
# Чётность функции
g1 = y(x) - y(-x)
g2 = -y(x) - y(-x)
if g1 == 0 and g2 != 0:
    show("Функция четная")
if g2 == 0 and g1 != 0:
    show("Функция нечетная")
else:
    show("Функция ни четная, ни нечетная")
```

<h3>2.2 Переодичность функции:</h3>

```python
var("T")
g = y(x) - y(T + x)
g.simplify()
solve(g, T)
```

Так как $T = 0$, следует, что функция не переодичная

<h3>7. Непрерывность. Наличие точек разрыва и их классификация:</h3>

```python
root = [0, 1]
for a in root:
    lim_left = lim(y, x = a, dir = 'left')
    lim_right = lim(y, x = a, dir = 'right')
    if (lim_left == lim_right and abs(lim_left) != oo and abs(lim_right) != oo):
        show("x = " + str(a) + " разрыв первого рода")
    else:
        show("x = " + str(a) + " разрыв второго рода")
```
