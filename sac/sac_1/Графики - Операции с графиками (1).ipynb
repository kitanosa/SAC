```python
# графики можно объединять:
x = var('x')
p1 = parametric_plot((cos(x),sin(x)),(x,0,2*pi),rgbcolor=hue(0.2))
p2 = parametric_plot((cos(x),sin(x)^2),(x,0,2*pi),rgbcolor=hue(0.4))
p3 = parametric_plot((cos(x),sin(x)^3),(x,0,2*pi),rgbcolor=hue(0.6))
show(p1+p2+p3, axes=false)
```

```python
x = var('x')
p1 = parametric_plot((cos(x),sin(x)),(x,0,2*pi),rgbcolor=hue(0.2))
p2 = parametric_plot((cos(x),sin(x)^2),(x,0,2*pi),rgbcolor=hue(0.4))
p3 = parametric_plot((cos(x),sin(x)^3),(x,0,2*pi),rgbcolor=hue(0.6))
p4 = parametric_plot((3*cos(x),sin(x)),(x,0,2*pi),rgbcolor=hue(0.1))
graphics_array([[p1, p2], [p3, p4]]).show(frame=True, axes=False)
```

```python
# с изменением размеров графика

L = [plot(sin(k*x),(x,-pi,pi)) for k in [1..3]]
G = graphics_array(L)
G.show(figsize=[5,3])  # smallish and compact
G.show(figsize=[10,20])  # bigger and tall and thin; long time (2s on sage.math, 2012)
G.show(figsize=8)  # figure as a whole is a square
```

```python
x = var('x')
p1 = parametric_plot((cos(x),sin(x)),(x,0,2*pi),rgbcolor=hue(0.2)
                     , linestyle="-" # or "solid" -- as default
                     , legend_label='solid'
                    )
p2 = plot(cos(x),(x,0,2*pi),rgbcolor=hue(1)
                     , linestyle="--" #  or "dashed"
                     , legend_label='dashed'
                    )
p3 = parametric_plot((cos(x),sin(x)^3),(x,0,2*pi),rgbcolor=hue(0.6)
                    , linestyle="-." # or "dash dot"
                    , legend_label='dashdot'
                    )
p4 = parametric_plot((3*cos(x),sin(x)),(x,0,2*pi),rgbcolor=hue(0.1)
                    , linestyle=":" #or "dotted"
                    , legend_label='dotted'
                    )
graphics_array([[p1, p2], [p3, p4]]).show(frame=True, axes=False)
```

```python
print("Можно заметить, что dashed для parametric_plot имеет свойство сливаться в сплошную линию")
parametric_plot((cos(x),sin(x)^2),(x,0,2*pi),rgbcolor=hue(1)
                     , linestyle="--" #  or "dashed"
                    )
```
