<h2>Задание 1 (Модель "Хищник-Жертва")</h2>

```python
# Положительные константы математической модели - системы дифференциальных уравнений Лотки-Вольтерра
a = 3
c = 1
d = 1
b = [4, 3, 2]

color = ["blue", "green", "red"]
graph = plot([])

x, y = var("x, y")

for n in range(len(b)):
    system = [x*(a-b[n]*y), y*(-c+d*x)]
    P = desolve_odeint(system,ics=[2, 1],times=srange(0,5,0.001),dvars=[x, y])
    graph += line(P, color = color[n])
show(graph)
```

<h2>Задание 2 (Падение тела, брошенного под углом к горизонту)</h2>

```python
k = 0.01
g = 9.81
v0 = 10
a = pi/4
tmax = 1.41
h = 0

y1, y2, y3, y4, t = var("y1 y2 y3 y4 t")

system = [y2, -k*y2*sqrt(y2^2+y4^2), y4, -k*y4*sqrt(y2^2+y4^2)-g]

T=desolve_system_rk4(system,[y1, y2, y3, y4],ics=[0,0,v0*cos(a),0,v0*sin(a)],ivar=t,end_points=tmax, step = 0.001)

Q=[[i,l] for i,j,k,l,t in T]

show(list_plot(Q))
```
