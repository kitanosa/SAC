# Модель "Хищник-Жертва"

```python
var("x y")


lines = line([])

for b in range(4,1,-1):
    eq = [x * (3 - b * y), y * (-1 + 1 * x)] #задаем функцию
    solution = desolve_odeint(des = eq, ics = [2, 1], times = srange(0, 7, 0.1), dvars=[x, y]) #получаем решение ОДУ dvars - зависимые переменные    Обыкновенное дифференциальное уравнение, des-правые стороны системы, ics - первоначальные значения times - последовательность моментов времени, в которые решение должно быть найденным
    lines += line(solution, rgbcolor=Color(randrange(100)/100, 1, 1, space='hsv').rgb()) #прибавляем объект для графика
    
lines.show(aspect_ratio = 1)#отображаем график
```

# Падение тела брошенного под углом к горизонту

```python
k = 0.01 #сопротивление среды
g = 9.81 #ускорение свободного падения
v0 = 10 #начальная скорость
a = pi/4 #угол
tmax = 1.41
h = 0

y1, y2, y3, y4, t = var("y1 y2 y3 y4 t") #переменные

system = [y2, -k*y2*sqrt(y2^2+y4^2), y4, -k*y4*sqrt(y2^2+y4^2)-g] #функция

T=desolve_system_rk4(system,[y1, y2, y3, y4],ics=[0,0,v0*cos(a),0,v0*sin(a)],ivar=t,end_points=tmax, step = 0.001) #решение ОДУ
Q=[[i,l] for i,j,k,l,t in T]

show(list_plot(Q))
```
